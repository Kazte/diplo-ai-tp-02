```python
# saco oral porque tiene un solo valor
cols_to_drop = ['oral', "ID"]
```

```python
df = df.drop(columns=cols_to_drop)
df
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df, x='gender', hue='smoking')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.histplot(data=df, x='age', bins=30, kde=True, hue='smoking')
plt.show()
```

```python
# agrego el indice de masa corporal para colapsar height y weight en una sola columna
df["BMI"] = df["weight(kg)"] / (df["height(cm)"]/100) ** 2
```

```python
cols_to_boxplot = ['systolic', 'relaxation', 'fasting blood sugar', 'Cholesterol','triglyceride', 'HDL', 'LDL', 'hemoglobin', 'Urine protein','serum creatinine', 'AST', 'ALT', 'Gtp', "BMI"]
```

```python
for col in cols_to_boxplot:
    plt.figure(figsize=(8,4))
    sns.histplot(data=df, x=col, bins=30, kde=True, hue='smoking')
    plt.title(f'Boxplot of {col} by Smoking Status')
    plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(data=df, x='Cholesterol', hue='smoking')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(data=df, x='triglyceride', hue='smoking')
plt.show()
```

```python
drop_cols = [
    'height(cm)', 'weight(kg)', 'hearing(left)', 'hearing(right)',
]
```

```python
df = df.drop(columns=drop_cols)
```

```python
# para hacer esto voy a asumir que tanto gender como tartar siempre llegan con datos.
# gender como M o F y tartar como Y o N
df = pd.get_dummies(df, columns=['gender', 'tartar'], dtype=int)
df
```

```python
df.smoking.value_counts()
```

```python
print(f"percetage of smokers: {df.smoking.value_counts()[1] / df.shape[0] * 100:.2f}%\npercentage of non-smokers: {df.smoking.value_counts()[0] / df.shape[0] * 100:.2f}%")
```

```python
df_smokers = df[df.smoking == 1]
df_non_smokers = df[df.smoking == 0]
```

```python
df_smokers
```

```python
temp_df = pd.DataFrame(columns=['column', 'dtype', 'unique_values'])
for col in df_smokers.columns:
  coltype = df_smokers[col].dtype
  temp_df = pd.concat([temp_df, pd.DataFrame({
      'column': [col],
      'dtype': [coltype],
      'unique_values': [df_smokers[col].nunique()]
  })], ignore_index=True)
  
temp_df
```

```python
temp_df = pd.DataFrame(columns=['column', 'dtype', 'unique_values'])
for col in df_non_smokers.columns:
  coltype = df_non_smokers[col].dtype
  temp_df = pd.concat([temp_df, pd.DataFrame({
      'column': [col],
      'dtype': [coltype],
      'unique_values': [df_non_smokers[col].nunique()]
  })], ignore_index=True)
  
temp_df
```

```python
temp_df = pd.DataFrame(columns=['column', 'dtype', 'unique_values'])
for col in df.columns:
  coltype = df[col].dtype
  temp_df = pd.concat([temp_df, pd.DataFrame({
      'column': [col],
      'dtype': [coltype],
      'unique_values': [df[col].nunique()]
  })], ignore_index=True)
  
temp_df
```
