````python
import pandas as pd
from sklearn.metrics import classification_report
````

````python
df = pd.read_csv("./processed_data/smoking_prediction_eda.csv")
df
````

````python
X = df.drop(columns=['smoking'])
y = df['smoking'].copy()
````

````python
seed = 42
````

````python
from sklearn.model_selection import train_test_split
````

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=seed, stratify=y)
````

Elegi el modelo XGBoost.

````python
from xgboost import XGBClassifier
from sklearn.metrics import (
    accuracy_score
)
````

````python
params = {
    "n_estimators": 500,
    "eta": 0.03,
    "max_depth": 4,
    "min_child_weight": 7,
    "subsample": 0.6,
    "colsample_bytree": 0.6,
    "gamma": 0.3,
    "reg_alpha": 0.5,
    "reg_lambda": 2,
}

clf = XGBClassifier(**params)

clf.fit(X_train, y_train)

train_pred = clf.predict(X_train)
test_pred = clf.predict(X_test)

print("Train report:")
print(classification_report(y_train, train_pred))
print("-----------------------------------------------------")
print("Test report:")
print(classification_report(y_test, test_pred))
````

````python
accuracy_gap = accuracy_score(y_train, train_pred) - accuracy_score(y_test, test_pred)
print("Accuracy gap between train and test: ", accuracy_gap)
````

Tenemos un Gap de accuracy entre train y test muy bajo (0.018 = 1.8%), lo que indica que el modelo no estaria haciendo overfitting (O en todo caso, es uno bastante bajo).
